# Example application of AbsurdER to T4L Data

```python
import absurdER as aer
import numpy as np
```

Path to input data. Input datasets are expect to be provided eithe

```python
rex = f'data/same_ff/nmr.npy'
rmd = f'data/same_ff/md.npy'
```

Print class docstrings to see what are the expected inputs

```python
print(aER.AbsurdER.__init__.__doc__)
```

For level1, we build a toy model from the blocks in rex, so we don't provide any experimental errors. Those will be automatically computed in the class. In this way, we can construct the `AbsurdER` class with minimal inputs. Results will be stored in a file called `results.pkl`, and the range of fudge parameters used will be `theta=0,100,1000,10000`. 

```python
rw = aER.AbsurdER( rex, rmd )
```

The construction of the class shows us also some information of the dataset, such as the number of methyls, the number of rates, the number of blocks and the overall chi squared between the simulated and the experimental rates. Let's take a look at how experimental and simulated rates look with respect to each other. 

```python
print(aER.AbsurdER.plot_comparison.__doc__)
```

```python
rw.plot_comparison( 0 )
```

We can now run the reweighting against rate 1 (`r=0`). This will take a while. 

```python
print(aER.AbsurdER.reweight.__doc__)
```

```python
rw.reweight( 0 )
```

Let's take a look at the results of the reweighting. First, we plot the curve `phi_eff` vs `chi2` for each rate. 

```python
print(aER.AbsurdER.plot_phix2r.__doc__)
```

```python
for i in range(3):
    rw.plot_phix2r( i )
```

Let us also take a look at how the scatter plot has changed now that the rates have been reweighted. 

```python
opt_theta = 100
rw.plot_comparison( 0, opt_theta )
```

In the case where we want to simply analyze some results without running any reweighting, we can construct the class and then run the `load_results` function. 

```python
rw.load_results('results.pkl')
```

Alternatively to reweighting with respect to rate 0 we can also run the whole routine using `run`. In this case, the reweighting will be performed optimizing with respect to all rates separately and all rates simultaneously. 

```python
print(aER.AbsurdER.run.__doc__)
```

```python
rw.run()
```

The results of the manuscript have been obtained in slighlty different conditions. For all the different examples, we used a more extended range of `thetas` and ignored `Ala146` from the analysis (methyl 14 in the dataset). The following blocks will reproduce the scatter plots in Fig. 2,3 and 4 of Ref. [1]. The calculations in each block will take around 1h. Therefore, we also provide the precomputed results. To run the minimization anyway, uncomment `rw.reweight` and comment `rw.load_results`. 

```python
rex    = f'level1/a99ildn_500_synth_tstar.npy'
rmd    = f'level1/a99ildn_1497_md_tstar.npy'
out    = "level1_Rall"
labels = [ r'($D_z$)', r'($D_y$)', r'($3D_z^2-2$)' ]
thetas = np.concatenate( ( np.array([0, 50]), np.arange( 100, 1500, 100 ), np.arange( 1500, 8500, 500 ) ) )

rw = aER.AbsurdER( rex, rmd, out = out, thetas = thetas, idx = [14] )
#rw.reweight( -1 )
rw.load_results('results/level1_Ra.pkl')

opt_theta = 300
for i in range(rw.r):
    rw.plot_comparison( i, opt_theta, labels[i] )
```

```python
rex    = f'level2/a15ipq_300_synth_tipq.npy'
rmd    = f'level2/a99ildn_1497_md_tipq.npy'
out    = "level2_Rall"
labels = [ r'($D_z$)', r'($D_y$)', r'($3D_z^2-2$)' ]
thetas = np.concatenate( ( np.array([0, 50]), np.arange( 100, 1500, 100 ), np.arange( 1500, 8500, 500 ) ) )

rw = aER.AbsurdER( rex, rmd, out = out, thetas = thetas, idx = [14] )
#rw.reweight( -1 )
rw.load_results('results/level2_Ra.pkl')

opt_theta = thetas[-9]
for i in range(rw.r):
    rw.plot_comparison( i, opt_theta, labels[i] )
```

```python
rex    = 'level3/nmr_rates.npy'
eex    = 'level3/nmr_errors.npy'
rmd    = 'level3/a99ildn_1497_md_tnmr.npy'
out    = 'level3_Rall'
labels = [ r'($D_z$)', r'($D_y$)', r'($3D_z^2-2$)' ]
thetas = np.concatenate( ( np.array([0, 50]), np.arange( 100, 1500, 100 ), np.arange( 1500, 8500, 500 ) ) )

rw = aER.AbsurdER( rex, rmd, eex, out = out, thetas = thetas, idx = [14] )
#rw.reweight( -1 )
rw.load_results('results/level3_Ra.pkl')

opt_theta = thetas[-13]
for i in range(rw.r):
    rw.plot_comparison( i, opt_theta, labels[i] )
```
